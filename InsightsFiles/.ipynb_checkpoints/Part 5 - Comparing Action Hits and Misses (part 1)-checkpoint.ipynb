```python
import re
import unicodedata
import nltk
from nltk import FreqDist
import pandas as pd
from nltk.corpus import stopwords
additional_stopwords = ['york', 'redhaired', 'jackson', 'redneck', 
                       'chang', 'philip', 'borat', 'nai', 'collin', 
                       'carol', 'zakarweski', 'jay', 'bob', 'lara',
                       'jean', 'peter', 'warner', 'bros', 'mr', 'le',
                        'crowley', 'heelshire', 'john', 'mike', 'texas',
                       'martin', 'sr', 'toby', 'midland', 'standard',
                       'dolio', 'hideko', 'llorona', 'jack', 'ig', 'de',
                        'la', 'salle', 'ii', 'wooden', 'tremaine', 'riding',
                        'acre', 'robert', 'downey', 'jr', 'janets', 'potts',
                       'gwyneth', 'paltrow', 'nala', 'stone', 'king', 'genie',
                        'thanos', '21', 'underpants', 'schmidt', 'jenko', 
                        'arthur', 'mera', 'tony', 'leaf']
import matplotlib.pyplot as plt
```

```python
def basic_clean(text):
    wnl = nltk.stem.WordNetLemmatizer()
    stopwords = nltk.corpus.stopwords.words('english') + additional_stopwords
    text = (unicodedata.normalize('NFKD', text)
           .encode('ascii', 'ignore')
           .decode('utf-8', 'ignore')
           .lower())
    words = re.sub(r'[^\w\s]', '', text).split()
    return [wnl.lemmatize(word) for word in words if word not in stopwords]
```

```python
with open('spoiler_text_files/action_hits.txt', 'r') as file:
    action_hits = file.read()
```

```python
with open('spoiler_text_files/action_misses.txt', 'r') as file:
    action_misses = file.read()
```

```python
len(action_hits), len(action_misses)
```

```python
type(action_hits), type(action_misses)
```

```python
action_hits = basic_clean(action_hits)
```

```python
action_misses = basic_clean(action_misses)
```

```python
fdist_action_hits = FreqDist(action_hits)
fdist_action_misses = FreqDist(action_misses)
```

# Most common words

```python
fdist_action_hits.most_common(10)
```

```python
fdist_action_misses.most_common(10)
```

```python
plt.ion()
fdist_action_hits.plot(20, title='Action Hits - Common Words', cumulative=False)
fdist_action_misses.plot(20, title = 'Action Misses - Common Words', cumulative=False)
plt.ioff()
plt.show()
```

# Bigrams

```python
#action hits trigram & action misses trigram
(pd.Series(nltk.ngrams(action_hits, 2)).value_counts())[:15]
```

```python
(pd.Series(nltk.ngrams(action_misses, 2)).value_counts())[:15]
```

```python
action_hits_bigrams = (pd.Series(nltk.ngrams(action_hits, 2)).value_counts())[:12]
action_misses_bigrams = (pd.Series(nltk.ngrams(action_misses, 2)).value_counts())[:12]
```

```python
action_hits_bigrams.sort_values().plot.barh(color='blue', 
                                           width=.9, figsize=(12,8))
plt.title('Action Hits - Bigrams')
plt.ylabel('Bigram')
plt.xlabel('# of Occurrences')
```

```python
action_misses_bigrams.sort_values().plot.barh(color='blue', 
                                           width=.9, figsize=(12,8))
plt.title('Action Misses - Bigrams')
plt.ylabel('Bigram')
plt.xlabel('# of Occurrences')
```
